```python
# -*- coding: utf-8 -*-
"""
乳腺癌诊断数据 - 全面探索性数据分析(EDA) - 16组图表完整版
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy import stats
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms
import warnings
warnings.filterwarnings('ignore')

# ==================== 设置英文环境（彻底避免方框） ====================
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_palette("husl")
plt.style.use('seaborn-v0_8-whitegrid')

# 创建保存图表的目录
import os
if not os.path.exists('eda_output'):
    os.makedirs('eda_output')

print("="*80)
print("Breast Cancer Diagnosis Data - Exploratory Data Analysis (EDA)")
print("="*80)

# ==================== 1. 数据加载 ====================
df = pd.read_csv('威斯康星乳腺癌诊断数据集.csv')
feature_cols = [col for col in df.columns if col not in ['ID', 'diagnosis']]

print(f"\nDataset Info:")
print(f"  - Samples: {df.shape[0]}")
print(f"  - Features: {len(feature_cols)}")
print(f"  - Benign (B): {(df['diagnosis']=='B').sum()} samples")
print(f"  - Malignant (M): {(df['diagnosis']=='M').sum()} samples")

# ==================== 图1: 目标变量分布 ====================
diagnosis_counts = df['diagnosis'].value_counts()
diagnosis_percent = df['diagnosis'].value_counts(normalize=True) * 100

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
colors_diag = ['#2ecc71', '#e74c3c']

# 柱状图
axes[0].bar(['Benign', 'Malignant'], diagnosis_counts.values, 
            color=colors_diag, edgecolor='black', linewidth=1.5)
axes[0].set_title('Diagnosis Frequency', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(diagnosis_counts.values):
    axes[0].text(i, v + 5, f'{v}\n({diagnosis_percent.iloc[i]:.1f}%)', 
                 ha='center', fontsize=12)

# 饼图
axes[1].pie(diagnosis_counts.values, labels=['Benign', 'Malignant'], autopct='%1.1f%%',
            colors=colors_diag, startangle=90, explode=(0.05, 0),
            wedgeprops={'edgecolor': 'black', 'linewidth': 1})
axes[1].set_title('Diagnosis Proportion', fontsize=14, fontweight='bold')

# 环形图
wedges, texts, autotexts = axes[2].pie(diagnosis_counts.values, 
                                        labels=['Benign', 'Malignant'], 
                                        autopct='%1.1f%%',
                                        colors=colors_diag, 
                                        startangle=90, 
                                        pctdistance=0.85,
                                        wedgeprops={'edgecolor': 'black', 'linewidth': 1})
centre_circle = plt.Circle((0,0), 0.70, fc='white')
axes[2].add_artist(centre_circle)
axes[2].set_title('Diagnosis Proportion (Donut)', fontsize=14, fontweight='bold')

plt.suptitle('Figure 1: Diagnosis Distribution Overview', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('eda_output/01_diagnosis_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 1 saved: 01_diagnosis_distribution.png")

# ==================== 图2: 雷达图 - 均值对比 ====================
selected_for_radar = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
                      'smoothness_mean', 'compactness_mean', 'concavity_mean', 
                      'concave_points_mean', 'symmetry_mean']

benign_means = [df[df['diagnosis']=='B'][f].mean() for f in selected_for_radar]
malignant_means = [df[df['diagnosis']=='M'][f].mean() for f in selected_for_radar]

# 归一化
max_vals = [max(df[f].max() for f in selected_for_radar)] * len(selected_for_radar)
benign_norm = [b/m for b, m in zip(benign_means, max_vals)]
malignant_norm = [m/max_v for m, max_v in zip(malignant_means, max_vals)]

angles = np.linspace(0, 2*np.pi, len(selected_for_radar), endpoint=False).tolist()
benign_norm += benign_norm[:1]
malignant_norm += malignant_norm[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))
ax.plot(angles, benign_norm, 'o-', linewidth=2, label='Benign', color='#2ecc71')
ax.fill(angles, benign_norm, alpha=0.25, color='#2ecc71')
ax.plot(angles, malignant_norm, 'o-', linewidth=2, label='Malignant', color='#e74c3c')
ax.fill(angles, malignant_norm, alpha=0.25, color='#e74c3c')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(selected_for_radar, size=9, rotation=45)
ax.set_ylim(0, 1)
ax.set_title('Figure 2: Feature Mean Comparison Radar Chart (Normalized)', 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
plt.tight_layout()
plt.savefig('eda_output/02_radar_chart.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 2 saved: 02_radar_chart.png")

# ==================== 图3: 特征分布综合分析 ====================
key_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
                'smoothness_mean', 'compactness_mean', 'concavity_mean', 
                'concave_points_mean', 'symmetry_mean']

fig = plt.figure(figsize=(20, 24))
for i, feature in enumerate(key_features):
    # 直方图+密度曲线
    ax1 = plt.subplot(9, 3, 3*i + 1)
    benign_data = df[df['diagnosis']=='B'][feature]
    malignant_data = df[df['diagnosis']=='M'][feature]
    
    ax1.hist(benign_data, bins=30, alpha=0.6, label='Benign', color='#2ecc71', density=True)
    ax1.hist(malignant_data, bins=30, alpha=0.6, label='Malignant', color='#e74c3c', density=True)
    benign_data.plot.kde(ax=ax1, color='#2ecc71', linewidth=2, linestyle='--')
    malignant_data.plot.kde(ax=ax1, color='#e74c3c', linewidth=2, linestyle='--')
    ax1.set_title(f'{feature} - Distribution', fontsize=10, fontweight='bold')
    ax1.set_xlabel('Value'); ax1.set_ylabel('Density')
    ax1.legend(fontsize=8)
    
    # 箱线图
    ax2 = plt.subplot(9, 3, 3*i + 2)
    bp = ax2.boxplot([benign_data, malignant_data], labels=['Benign', 'Malignant'],
                     patch_artist=True, widths=0.6)
    bp['boxes'][0].set_facecolor('#2ecc71')
    bp['boxes'][1].set_facecolor('#e74c3c')
    bp['medians'][0].set_color('black')
    bp['medians'][1].set_color('black')
    ax2.set_title(f'{feature} - Boxplot', fontsize=10, fontweight='bold')
    ax2.set_ylabel('Value')
    
    # Q-Q图
    ax3 = plt.subplot(9, 3, 3*i + 3)
    stats.probplot(df[feature], dist="norm", plot=ax3)
    ax3.set_title(f'{feature} - Q-Q Plot', fontsize=10, fontweight='bold')

plt.suptitle('Figure 3: Comprehensive Feature Distribution Analysis\n(Histogram + Density + Boxplot + Q-Q Plot)', 
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.savefig('eda_output/03_distribution_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 3 saved: 03_distribution_analysis.png")

# ==================== 图4: 全特征箱线图 ====================
n_features = len(feature_cols)
n_cols = 4
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 4*n_rows))
axes = axes.flatten()

for i, feature in enumerate(feature_cols):
    data_to_plot = [df[df['diagnosis']=='B'][feature], 
                    df[df['diagnosis']=='M'][feature]]
    bp = axes[i].boxplot(data_to_plot, labels=['B', 'M'], patch_artist=True,
                          boxprops=dict(linewidth=1.2),
                          medianprops=dict(linewidth=2, color='red'),
                          whiskerprops=dict(linewidth=1.2),
                          capprops=dict(linewidth=1.2))
    bp['boxes'][0].set_facecolor('#2ecc71')
    bp['boxes'][1].set_facecolor('#e74c3c')
    axes[i].set_title(feature, fontsize=8, fontweight='bold')
    axes[i].tick_params(axis='x', labelsize=7)

for i in range(len(feature_cols), len(axes)):
    axes[i].set_visible(False)

plt.suptitle('Figure 4: Boxplots of All Features (Benign vs Malignant)', 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_output/04_all_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 4 saved: 04_all_boxplots.png")

# ==================== 图5: 异常值统计 ====================
outlier_results = []
for col in feature_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_results.append({
        'Feature': col,
        'Outlier Count': len(outliers),
        'Outlier Pct(%)': len(outliers) / len(df) * 100
    })

outlier_df = pd.DataFrame(outlier_results)
outlier_df = outlier_df.sort_values('Outlier Pct(%)', ascending=False)

fig, ax = plt.subplots(figsize=(14, 8))
top_outliers = outlier_df.head(20)
colors_out = ['#e74c3c' if x > 5 else '#f39c12' if x > 2 else '#2ecc71' 
              for x in top_outliers['Outlier Pct(%)']]
bars = ax.barh(range(len(top_outliers)), top_outliers['Outlier Pct(%)'], color=colors_out)
ax.set_yticks(range(len(top_outliers)))
ax.set_yticklabels(top_outliers['Feature'])
ax.set_xlabel('Outlier Percentage (%)', fontsize=12)
ax.set_title('Figure 5: Outlier Percentage by Feature', fontsize=14, fontweight='bold')
ax.axvline(x=5, color='red', linestyle='--', linewidth=1, label='5% threshold')
ax.axvline(x=2, color='orange', linestyle='--', linewidth=1, label='2% threshold')
ax.legend()
plt.tight_layout()
plt.savefig('eda_output/05_outlier_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 5 saved: 05_outlier_analysis.png")

# ==================== 图6: 全特征热力图 ====================
corr_matrix = df[feature_cols].corr()

fig, ax = plt.subplots(figsize=(20, 18))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
sns.heatmap(corr_matrix, mask=mask, cmap='RdBu_r', center=0, 
            annot=False, square=True, linewidths=0.3,
            cbar_kws={"shrink": 0.8, "label": "Correlation"}, ax=ax)
ax.set_title('Figure 6: Full Feature Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_output/06_full_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 6 saved: 06_full_heatmap.png")

# ==================== 图7: 聚类热力图 ====================
fig = plt.figure(figsize=(16, 14))
sns.clustermap(corr_matrix, cmap='RdBu_r', center=0, 
               figsize=(16, 14), dendrogram_ratio=0.1,
               cbar_pos=(0.02, 0.8, 0.03, 0.18))
plt.suptitle('Figure 7: Clustered Correlation Heatmap', fontsize=14, fontweight='bold', y=1.02)
plt.savefig('eda_output/07_clustered_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 7 saved: 07_clustered_heatmap.png")

# ==================== 图8: 相关性条形图 ====================
diagnosis_numeric = df['diagnosis'].map({'B': 0, 'M': 1})
corr_with_diag = df[feature_cols].apply(lambda x: x.corr(diagnosis_numeric))
corr_with_diag_sorted = corr_with_diag.sort_values()

fig, ax = plt.subplots(figsize=(12, 10))
colors_corr = ['#e74c3c' if x > 0 else '#2ecc71' for x in corr_with_diag_sorted.values]
bars = ax.barh(corr_with_diag_sorted.index, corr_with_diag_sorted.values, color=colors_corr)
ax.set_xlabel('Correlation Coefficient', fontsize=12)
ax.set_title('Figure 8: Feature Correlation with Diagnosis\n(Red=Positive→Malignant, Green=Negative→Benign)', 
             fontsize=12, fontweight='bold')
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax.axvline(x=0.5, color='red', linestyle='--', linewidth=0.8, alpha=0.5)
ax.axvline(x=-0.5, color='green', linestyle='--', linewidth=0.8, alpha=0.5)
ax.set_xlim(-1, 1)
plt.tight_layout()
plt.savefig('eda_output/08_correlation_barchart.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 8 saved: 08_correlation_barchart.png")

# ==================== 图9: 小提琴图对比 ====================
selected_violin = ['radius_mean', 'area_mean', 'concavity_mean', 'concave_points_mean',
                   'radius_worst', 'area_worst', 'concavity_worst', 'concave_points_worst']

fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for i, feature in enumerate(selected_violin):
    data_to_plot = [df[df['diagnosis']=='B'][feature].values,
                    df[df['diagnosis']=='M'][feature].values]
    parts = axes[i].violinplot(data_to_plot, positions=[1, 2], showmeans=True, showmedians=True)
    parts['bodies'][0].set_facecolor('#2ecc71')
    parts['bodies'][0].set_alpha(0.7)
    parts['bodies'][1].set_facecolor('#e74c3c')
    parts['bodies'][1].set_alpha(0.7)
    
    axes[i].set_xticks([1, 2])
    axes[i].set_xticklabels(['Benign', 'Malignant'])
    axes[i].set_title(feature, fontsize=11, fontweight='bold')
    axes[i].set_ylabel('Value')
    
    # 添加均值文本
    mean_b = df[df['diagnosis']=='B'][feature].mean()
    mean_m = df[df['diagnosis']=='M'][feature].mean()
    axes[i].text(1, mean_b, f'μ={mean_b:.2f}', ha='center', va='bottom', fontsize=8)
    axes[i].text(2, mean_m, f'μ={mean_m:.2f}', ha='center', va='bottom', fontsize=8)

plt.suptitle('Figure 9: Violin Plots of Core Features (Benign vs Malignant)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_output/09_violin_plots.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 9 saved: 09_violin_plots.png")

# ==================== 图10: 散点图矩阵 ====================
selected_scatter = key_features[:5]
colors_scatter = ['#2ecc71' if d == 'B' else '#e74c3c' for d in df['diagnosis']]

from pandas.plotting import scatter_matrix
fig = scatter_matrix(df[selected_scatter], figsize=(15, 15), alpha=0.6,
                      c=colors_scatter, diagonal='hist')
for ax in fig.ravel():
    ax.set_xlabel(ax.get_xlabel(), fontsize=8)
    ax.set_ylabel(ax.get_ylabel(), fontsize=8)
plt.suptitle('Figure 10: Scatter Matrix of Key Features (Green=Benign, Red=Malignant)', 
             fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout()
plt.savefig('eda_output/10_scatter_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 10 saved: 10_scatter_matrix.png")

# ==================== 图11: PCA碎石图 ====================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[feature_cols])
y = df['diagnosis'].map({'B': 0, 'M': 1})

pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_scaled)
explained_variance = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 碎石图
axes[0].plot(range(1, len(explained_variance)+1), explained_variance, 
             'o-', linewidth=2, markersize=6, color='#3498db')
axes[0].set_xlabel('Principal Component', fontsize=12)
axes[0].set_ylabel('Explained Variance Ratio', fontsize=12)
axes[0].set_title('Figure 11a: Scree Plot', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# 累计方差图
axes[1].plot(range(1, len(cumulative_variance)+1), cumulative_variance, 
             'o-', linewidth=2, markersize=6, color='#e74c3c')
axes[1].set_xlabel('Principal Component', fontsize=12)
axes[1].set_ylabel('Cumulative Variance Ratio', fontsize=12)
axes[1].set_title('Figure 11b: Cumulative Variance', fontsize=12, fontweight='bold')
axes[1].axhline(y=0.95, color='green', linestyle='--', label='95% threshold')
axes[1].axhline(y=0.90, color='orange', linestyle='--', label='90% threshold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('eda_output/11_pca_scree.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 11 saved: 11_pca_scree.png")

# ==================== 图12: PCA二维散点图 ====================
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 基本散点图
axes[0].scatter(X_pca_2d[y==0, 0], X_pca_2d[y==0, 1], 
                c='#2ecc71', label='Benign', alpha=0.7, edgecolors='black', s=50)
axes[0].scatter(X_pca_2d[y==1, 0], X_pca_2d[y==1, 1], 
                c='#e74c3c', label='Malignant', alpha=0.7, edgecolors='black', s=50)
axes[0].set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
axes[0].set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
axes[0].set_title('Figure 12a: PCA 2D Scatter Plot', fontsize=12, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

def confidence_ellipse(x, y, ax, n_std=2.0, facecolor='none', **kwargs):
    if x.size != y.size:
        raise ValueError("x and y must be the same size")
    cov = np.cov(x, y)
    pearson = cov[0, 1] / np.sqrt(cov[0, 0] * cov[1, 1])
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0), width=ell_radius_x * 2, height=ell_radius_y * 2,
                      facecolor=facecolor, **kwargs)
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = np.mean(x)
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = np.mean(y)
    transf = transforms.Affine2D().rotate_deg(45).scale(scale_x, scale_y).translate(mean_x, mean_y)
    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)

# 带置信椭圆
axes[1].scatter(X_pca_2d[y==0, 0], X_pca_2d[y==0, 1], 
                c='#2ecc71', alpha=0.5, s=40, label='Benign')
axes[1].scatter(X_pca_2d[y==1, 0], X_pca_2d[y==1, 1], 
                c='#e74c3c', alpha=0.5, s=40, label='Malignant')
confidence_ellipse(X_pca_2d[y==0, 0], X_pca_2d[y==0, 1], axes[1], 
                   n_std=2, edgecolor='#2ecc71', linewidth=2, alpha=0.8)
confidence_ellipse(X_pca_2d[y==1, 0], X_pca_2d[y==1, 1], axes[1], 
                   n_std=2, edgecolor='#e74c3c', linewidth=2, alpha=0.8)
axes[1].set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
axes[1].set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
axes[1].set_title('Figure 12b: PCA 2D with 95% Confidence Ellipse', fontsize=12, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('eda_output/12_pca_2d.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 12 saved: 12_pca_2d.png")

# ==================== 图13: PCA三维散点图 ====================
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca_3d[y==0, 0], X_pca_3d[y==0, 1], X_pca_3d[y==0, 2],
           c='#2ecc71', label='Benign', alpha=0.6, s=30)
ax.scatter(X_pca_3d[y==1, 0], X_pca_3d[y==1, 1], X_pca_3d[y==1, 2],
           c='#e74c3c', label='Malignant', alpha=0.6, s=30)
ax.set_xlabel(f'PC1 ({pca_3d.explained_variance_ratio_[0]*100:.1f}%)', fontsize=10)
ax.set_ylabel(f'PC2 ({pca_3d.explained_variance_ratio_[1]*100:.1f}%)', fontsize=10)
ax.set_zlabel(f'PC3 ({pca_3d.explained_variance_ratio_[2]*100:.1f}%)', fontsize=10)
ax.set_title('Figure 13: PCA 3D Scatter Plot', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig('eda_output/13_pca_3d.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 13 saved: 13_pca_3d.png")

# ==================== 图14: PCA载荷图 ====================
pca_full_2d = PCA(n_components=2)
pca_full_2d.fit(X_scaled)
loadings = pca_full_2d.components_.T

fig, ax = plt.subplots(figsize=(12, 10))
# 绘制载荷向量
for i, feature in enumerate(feature_cols):
    ax.arrow(0, 0, loadings[i, 0], loadings[i, 1], 
             head_width=0.03, head_length=0.03, fc='blue', ec='blue', alpha=0.6)
    ax.text(loadings[i, 0]*1.05, loadings[i, 1]*1.05, feature, fontsize=7)

# 绘制数据点
ax.scatter(X_pca_2d[y==0, 0], X_pca_2d[y==0, 1], c='#2ecc71', alpha=0.3, s=20, label='Benign')
ax.scatter(X_pca_2d[y==1, 0], X_pca_2d[y==1, 1], c='#e74c3c', alpha=0.3, s=20, label='Malignant')

# 添加单位圆
circle = plt.Circle((0, 0), 1, fill=False, linestyle='--', color='gray')
ax.add_patch(circle)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_xlabel(f'PC1 ({pca_full_2d.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
ax.set_ylabel(f'PC2 ({pca_full_2d.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
ax.set_title('Figure 14: PCA Loading Plot (Feature Contributions)', fontsize=14, fontweight='bold')
ax.axhline(y=0, color='gray', linestyle='-', linewidth=0.5)
ax.axvline(x=0, color='gray', linestyle='-', linewidth=0.5)
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig('eda_output/14_pca_loadings.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 14 saved: 14_pca_loadings.png")

# ==================== 图15: 偏度和峰度分析 ====================
skewness = df[feature_cols].skew()
kurtosis = df[feature_cols].kurtosis()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 偏度图
skewness_sorted = skewness.sort_values()
colors_skew = ['#e74c3c' if abs(x) > 1 else '#f39c12' if abs(x) > 0.5 else '#2ecc71' 
               for x in skewness_sorted.values]
axes[0].barh(range(len(skewness_sorted)), skewness_sorted.values, color=colors_skew)
axes[0].set_yticks(range(len(skewness_sorted)))
axes[0].set_yticklabels(skewness_sorted.index)
axes[0].set_xlabel('Skewness', fontsize=12)
axes[0].set_title('Figure 15a: Feature Skewness Analysis\n(Normal distribution skewness ≈ 0)', 
                  fontsize=12, fontweight='bold')
axes[0].axvline(x=0, color='black', linestyle='-', linewidth=0.8)
axes[0].axvline(x=0.5, color='orange', linestyle='--', alpha=0.5)
axes[0].axvline(x=-0.5, color='orange', linestyle='--', alpha=0.5)

# 峰度图
kurtosis_sorted = kurtosis.sort_values()
colors_kurt = ['#e74c3c' if x > 3 else '#f39c12' if x > 1 else '#2ecc71' 
               for x in kurtosis_sorted]
axes[1].barh(range(len(kurtosis_sorted)), kurtosis_sorted.values, color=colors_kurt)
axes[1].set_yticks(range(len(kurtosis_sorted)))
axes[1].set_yticklabels(kurtosis_sorted.index)
axes[1].set_xlabel('Kurtosis', fontsize=12)
axes[1].set_title('Figure 15b: Feature Kurtosis Analysis\n(Normal distribution kurtosis ≈ 0, Leptokurtic > 0)', 
                  fontsize=12, fontweight='bold')
axes[1].axvline(x=0, color='black', linestyle='-', linewidth=0.8)

plt.tight_layout()
plt.savefig('eda_output/15_skewness_kurtosis.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 15 saved: 15_skewness_kurtosis.png")

# ==================== 图16: 核心特征小提琴图汇总 ====================
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for i, feature in enumerate(selected_violin):
    data_to_plot = [df[df['diagnosis']=='B'][feature].values,
                    df[df['diagnosis']=='M'][feature].values]
    parts = axes[i].violinplot(data_to_plot, positions=[1, 2], showmeans=True, showmedians=True)
    parts['bodies'][0].set_facecolor('#2ecc71')
    parts['bodies'][0].set_alpha(0.7)
    parts['bodies'][1].set_facecolor('#e74c3c')
    parts['bodies'][1].set_alpha(0.7)
    axes[i].set_xticks([1, 2])
    axes[i].set_xticklabels(['Benign', 'Malignant'])
    axes[i].set_title(feature, fontsize=11, fontweight='bold')
    axes[i].set_ylabel('Value')

plt.suptitle('Figure 16: Core Features Violin Plot Summary', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_output/16_core_features_violin.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Figure 16 saved: 16_core_features_violin.png")

# ==================== 总结 ====================
print("\n" + "="*80)
print("EDA COMPLETED SUCCESSFULLY!")
print("="*80)
print("\nAll 16 figures have been saved to 'eda_output' directory:")
print("")
print("  FIG 1: 01_diagnosis_distribution.png      - Diagnosis distribution (bar/pie/donut)")
print("  FIG 2: 02_radar_chart.png                 - Radar chart of feature means")
print("  FIG 3: 03_distribution_analysis.png       - Histogram + Density + Boxplot + Q-Q")
print("  FIG 4: 04_all_boxplots.png                - Boxplots of all 30 features")
print("  FIG 5: 05_outlier_analysis.png            - Outlier percentage statistics")
print("  FIG 6: 06_full_heatmap.png                - Full correlation heatmap")
print("  FIG 7: 07_clustered_heatmap.png           - Clustered correlation heatmap")
print("  FIG 8: 08_correlation_barchart.png        - Correlation with diagnosis")
print("  FIG 9: 09_violin_plots.png                - Violin plots (8 core features)")
print("  FIG10: 10_scatter_matrix.png              - Scatter matrix of key features")
print("  FIG11: 11_pca_scree.png                   - PCA scree & cumulative variance")
print("  FIG12: 12_pca_2d.png                      - PCA 2D with confidence ellipse")
print("  FIG13: 13_pca_3d.png                      - PCA 3D scatter plot")
print("  FIG14: 14_pca_loadings.png                - PCA loading plot")
print("  FIG15: 15_skewness_kurtosis.png           - Skewness & kurtosis analysis")
print("  FIG16: 16_core_features_violin.png        - Core features violin summary")
print("")
print("="*80)
```
